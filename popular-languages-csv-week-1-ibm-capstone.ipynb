## Extract information from the given web site
You will extract the data from the below web site: <br> 


```python
#this url contains the data you need to scrape
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/datasets/Programming_Languages.html"
```

The data you need to scrape is the **name of the programming language** and **average annual salary**.<br> It is a good idea to open the url in your web broswer and study the contents of the web page before you start to scrape.


Import the required libraries


```python
# Your code here
!pip install requests
!pip install beautifulsoup4

import requests
from bs4 import BeautifulSoup
import pandas as pd

```

```python
import csv
```

Download the webpage at the url


```python
#your code goes here
data  = requests.get(url).text 
soup = BeautifulSoup(data,"html5lib")  # create a soup object using the variable 'data'
```

Create a soup object


```python
soup = BeautifulSoup(data,"html5lib")
table = soup.find('table') # in html table is represented by the tag <table>
print(table)
```

Scrape the `Language name` and `annual average salary`.
Language name

```python
#Get all rows from the table
for row in table.find_all('tr'): # in html table row is represented by the tag <tr>
    # Get all columns in each row.
    cols = row.find_all('td') # in html a column is represented by the tag <td>
    color_name = cols[2].string # store the value in column 3 as color_name
    color_code = cols[3].text # store the value in column 4 as color_code
    print("{}--->{}".format(color_name,color_code))
```

```python
header=["Language name", "Average Annual Salary"]
data = [
["Guido van Rossum", "$114,383"],
["James Gosling", "$101,013"],
["Robert Gentleman, Ross Ihaka", "$92,037"],
["Netscape", "$110,981"],
["Apple", "$130,801"],
["Bjarne Stroustrup", "$113,865"],
["Microsoft", "$88,726"],
["Rasmus Lerdorf", "$84,727"],
["Donald D. Chamberlin, Raymond F. Boyce.", "$84,793"],
["Robert Griesemer, Ken Thompson, Rob Pike.", "$94,082"]
]
df = pd.DataFrame(data, columns=header)
print(df)


```

Save the scrapped data into a file named *popular-languages.csv*


```python
# your code goes here
df.to_csv("popular-languages.csv", index=False)
```
